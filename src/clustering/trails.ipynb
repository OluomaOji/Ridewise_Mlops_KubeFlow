k-mean clustering 


```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Select the features for clustering
features = df[['pickup_lat', 'pickup_lng', 'dropoff_lat', 'dropoff_lng', 'fare', 'tip']]
```

```python
# Handle missing values
features = features.dropna()
```

```python
# standardize the features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

```python
# apply elbow method to find the optimal number of clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42, init='k-means++')
    kmeans.fit(features_scaled)
    wcss.append(kmeans.inertia_)
```

```python
# plot the elbow method
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method for Optimal Clusters')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.xticks(range(1, 11))
plt.grid()
plt.show()
```

```python
# Set the optimal number of clusters (from elbow method, e.g., i=10)
optimal_clusters = 4 # or set to the value you choose, e.g., 10

# initial model
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42, init='k-means++')
cluster = kmeans.fit_predict(features_scaled)
cluster
```

```python
# Plot clusters using pickup lat and lng and fare from the features DataFrame
plt.figure(figsize=(10, 6))
scatter = plt.scatter(features['pickup_lng'], features['pickup_lat'], s=features['fare'],  # Use fare as the size of the points
                      c=cluster, cmap='viridis', alpha=0.6, edgecolor='k')

plt.xlabel('Pickup Longitude')
plt.ylabel('Pickup Latitude')
plt.title('K-Means Clusters of Trip Pickup Locations')
plt.colorbar(scatter, label='Cluster')
plt.grid(True)
plt.tight_layout()
plt.savefig('kmeans_clusters.png', dpi=300, bbox_inches='tight')
plt.show()
```

DBSCAN CLUSTERING 

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
```

```python
features
```

```python
dbscan = DBSCAN(eps=0.3, min_samples=10)  # You can tune these values
dbscan.fit_predict(features_scaled)
```

```python
# count the number of clusters
n_clusters = len(set(dbscan.labels_)) - (1 if -1 in dbscan.labels_ else 0)
n_noise = list(dbscan.labels_).count(-1)
```

```python
# print the number of clusters and noise points
print(f"Number of clusters: {n_clusters}")
print(f"Number of noise points: {n_noise}")
```

```python
# Plot the clusters
plt.figure(figsize=(10, 6))
unique_labels = set(dbscan.labels_)

for label in unique_labels:
    color = 'k' if label == -1 else plt.cm.jet(float(label) / max(unique_labels))
    class_member_mask = (dbscan.labels_ == label)
    xy = features[class_member_mask]
    plt.plot(xy['pickup_lng'], xy['pickup_lat'], 'o', markerfacecolor=color,
             markeredgecolor='k', markersize=7)

plt.title('DBSCAN Clustering of Pickup Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.savefig('dbscan_clusters.png', dpi=400, bbox_inches='tight')
plt.show()
```

```python
from sklearn.manifold import TSNE

# Run t-SNE on the standardized data
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
X_tsne = tsne.fit_transform(features_scaled)

# Plot t-SNE results with cluster labels
plt.figure(figsize=(10, 6))
for label in set(dbscan.labels_):
    mask = dbscan.labels_ == label
    plt.scatter(X_tsne[mask, 0], X_tsne[mask, 1],
                label=f'Cluster {label}' if label != -1 else 'Noise',
                alpha=0.7, edgecolor='k')

plt.title("DBSCAN Clustering Visualized with t-SNE")
plt.legend()
plt.grid(True)
plt.show()
plt.savefig('tsne.png', dpi=400, bbox_inches='tight')

```

```python
from sklearn.decomposition import PCA

# Run PCA on the standardized data
pca = PCA(n_components=2)
X_pca = pca.fit_transform(features_scaled)

# Plot PCA results with cluster labels
plt.figure(figsize=(10, 6))
for label in set(dbscan.labels_):
    mask = dbscan.labels_ == label
    plt.scatter(X_pca[mask, 0], X_pca[mask, 1],
                label=f'Cluster {label}' if label != -1 else 'Noise',
                alpha=0.7, edgecolor='k')

plt.title("DBSCAN Clustering Visualized with PCA")
plt.legend()
plt.grid(True)
plt.savefig('pca.png', dpi=400, bbox_inches='tight')
plt.show()
```
